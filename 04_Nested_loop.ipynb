# Prime or not Prime

```python
n= 21
c= 0

for i in range (1,n+1):
    if n%i== 0:
        c+= 1
if c == 2:
    print('Prime')
else:
    print('Not Prime')
```

```python
n = 9
prime = True

for i in range(2, n):  # Check from 2 to n-1
    if n % i == 0:
        prime = False
        break

if prime and n > 1:
    print("Prime")
else:
    print("Not Prime")
        
```

# First 100 Prime numbers

```python
count = 0
n = 2  # Start from the first prime number

while count < 100:
    prime = True
    for i in range(2, int(n**0.5) + 1):  # More efficient
        if n % i == 0:
            prime = False
            break

    if prime:
        print(n)
        count += 1
    n += 1
```

# getting a specific number

```python
c=0
for i in range (1,7):
    for j in range(1,7):
        if i+j == 8:
            print(i,j)
        c+= 1
        
print(c)        
```
